```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('forestfires.csv')
```

```python
df
```

```python
#a. Scatter Plot
plt.scatter(df['month'], df['wind'])
# Adding Title to the Plot
plt.title("Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Month')
plt.ylabel('wind')
plt.show()
```

```python
#b. Bar Plot
#Bar chart with day against temprature
plt.bar(df['day'], df['temp'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Temprature')
# Adding the legends
plt.show()
```

```python
#c. Histogram
# histogram of total_bills
plt.hist(df['DMC'])
plt.title("Histogram")
plt.show()
```
