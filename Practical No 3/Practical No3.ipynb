```python
#Reading and Removing Duplicate Values
df.duplicated(subset=['CO(GT)'])
```

```python
#Remove Duplicates
df.drop_duplicates(keep=False)
```

```python
#Handling Outliers:
def remove_outliers(df, columns, n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        mean = df[col].mean()
        sd = df[col].std()
        df = df[(df[col] <= mean + (n_std * sd))]
    return df
df
```
